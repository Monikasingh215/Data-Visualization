```python
# ecommerce_data_analysis.py

import mysql.connector
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Database Connectivity
def connect_to_database():
    connection = mysql.connector.connect(
        user='root',
        password='monika@123',
        host='localhost',
        database='ecommerce'
    )
    return connection

# Fetch data from the database
def fetch_data():
    connection = connect_to_database()
    cursor = connection.cursor()

    # Query data from the 'customer' table
    cursor.execute('SELECT * FROM customer')
    customer_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

    # Query data from the 'product' table
    cursor.execute('SELECT * FROM product')
    product_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

    # Query data from the 'order_details' table
    cursor.execute('SELECT * FROM order_details')
    order_data = pd.DataFrame(cursor.fetchall(), columns=[desc[0] for desc in cursor.description])

    cursor.close()
    connection.close()

    return customer_data, product_data, order_data

# Data Cleaning
def clean_data(customer_data, product_data, order_data):
    # Check for missing values
    print("Customer Data Missing Values:\n", customer_data.isnull().sum())
    print("Product Data Missing Values:\n", product_data.isnull().sum())
    print("Order Data Missing Values:\n", order_data.isnull().sum())

    # Ensure data types are appropriate
    print("Customer Data Types:\n", customer_data.dtypes)
    print("Product Data Types:\n", product_data.dtypes)
    print("Order Data Types:\n", order_data.dtypes)

    # Handle missing values if any (example: filling with a placeholder)
    customer_data.fillna('Unknown', inplace=True)
    product_data.fillna({'stock': 0}, inplace=True)
    order_data.fillna({'quantity': 0}, inplace=True)

    # Convert columns to appropriate data types
    customer_data['pin_code'] = customer_data['pin_code'].astype(str)
    order_data['order_date'] = pd.to_datetime(order_data['order_date'])

    return customer_data, product_data, order_data

# Exploratory Data Analysis (EDA)
def perform_eda(customer_data, product_data, order_data):
    sns.set(style="whitegrid")

    # Customer Analysis
    print("\nCustomer analysis :- \n")
    print("Total Number of Customers City-wise:")
    city_counts = customer_data['city'].value_counts()
    print(city_counts)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x=city_counts.index, y=city_counts.values, palette="viridis", hue=city_counts.index, dodge=False)
    plt.title("Total Number of Customers City-wise")
    plt.xlabel("City")
    plt.ylabel("Number of Customers")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    # Most frequent customers based on order history
    customer_order_counts = order_data['customer_id'].value_counts()
    frequent_customers = customer_data[customer_data['customer_id'].isin(customer_order_counts.head(10).index)]
    print("Most Frequent Customers:\n", frequent_customers)

    # Product Analysis
    print("\nProduct analysis :- \n")
    print("Total Number of Products Available by Category:")
    product_category_counts = product_data['category'].value_counts()
    print(product_category_counts)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=product_category_counts.index, y=product_category_counts.values, palette="plasma", hue=product_category_counts.index, dodge=False)
    plt.title("Total Number of Products by Category")
    plt.xlabel("Category")
    plt.ylabel("Number of Products")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    print("Distribution of Products Across Sub-categories:")
    product_sub_category_counts = product_data['sub_category'].value_counts()
    print(product_sub_category_counts)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=product_sub_category_counts.index, y=product_sub_category_counts.values, palette="magma", hue=product_sub_category_counts.index, dodge=False)
    plt.title("Distribution of Products Across Sub-categories")
    plt.xlabel("Sub-category")
    plt.ylabel("Number of Products")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    print("Products with Low Stock Levels:")
    low_stock_products = product_data[product_data['stock'] < 10]
    print(low_stock_products)

    print("Average, Maximum, and Minimum Prices for Products:")
    price_stats = product_data[['selling_price']].agg(['mean', 'max', 'min'])
    print(price_stats)

    plt.figure(figsize=(8, 6))
    sns.histplot(product_data['selling_price'], kde=True, color="blue")
    plt.title("Distribution of Product Selling Prices")
    plt.xlabel("Selling Price")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()

    # Order Analysis
    print("\nOrder analysis :- \n")
    print("Top 10 Orders Product Wise:")
    top_orders = order_data.groupby('product_id')['quantity'].sum().nlargest(10)
    print(top_orders)

    print("Order Status Distribution:")
    order_status_distribution = order_data['order_status'].value_counts()
    print(order_status_distribution)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=order_status_distribution.index, y=order_status_distribution.values, palette="coolwarm", hue=order_status_distribution.index, dodge=False)
    plt.title("Order Status Distribution")
    plt.xlabel("Order Status")
    plt.ylabel("Number of Orders")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    print("Most Popular Products Based on Order Quantity:")
    popular_products = order_data.groupby('product_id')['quantity'].sum().sort_values(ascending=False)
    print(popular_products.head(10))

    # Sales Analysis
    print("\nSales analysis :- \n")
    print("Total Revenue Generated from Orders Product Wise:")
    revenue_per_product = order_data.groupby('product_id')['total_price'].sum()
    print(revenue_per_product)

    print("Total Revenue Generated from All Orders:")
    total_revenue = order_data['total_price'].sum()
    print("Total Revenue: ", total_revenue)

    print("Total Revenue Product Category Wise Percentage:")
    revenue_per_category = order_data.merge(product_data, on='product_id').groupby('category')['total_price'].sum()
    category_percentage = (revenue_per_category / total_revenue) * 100
    print(category_percentage)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=category_percentage.index, y=category_percentage.values, palette="crest", hue=category_percentage.index, dodge=False)
    plt.title("Revenue Percentage by Product Category")
    plt.xlabel("Category")
    plt.ylabel("Revenue Percentage")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    print("Most Profitable Products Based on Difference Between Original and Selling Prices:")
    product_data['profit'] = product_data['original_price'] - product_data['selling_price']
    most_profitable_products = product_data.sort_values(by='profit', ascending=False)
    print(most_profitable_products.head(10))

    # Customer Order Patterns
    print("Product Names with Highest and Lowest Order Quantities:")
    order_quantities = order_data.groupby('product_id')['quantity'].sum()
    max_quantity_product_ids = order_quantities[order_quantities == order_quantities.max()].index.tolist()
    min_quantity_product_ids = order_quantities[order_quantities == order_quantities.min()].index.tolist()

    most_ordered_products = product_data[product_data['product_id'].isin(max_quantity_product_ids)]
    least_ordered_products = product_data[product_data['product_id'].isin(min_quantity_product_ids)]

    print("Highest Ordered Products:\n", most_ordered_products)
    print("Lowest Ordered Products:\n", least_ordered_products)

    print("Customers with Highest and Lowest Order Quantities by Name:")
    customer_order_totals = order_data.groupby('customer_id')['quantity'].sum()
    top_customers = customer_data[customer_data['customer_id'].isin(customer_order_totals[customer_order_totals == customer_order_totals.max()].index)]
    bottom_customers = customer_data[customer_data['customer_id'].isin(customer_order_totals[customer_order_totals == customer_order_totals.min()].index)]

    print("Top Customers:\n", top_customers)
    print("Bottom Customers:\n", bottom_customers)

    print("Most Preferred Payment Modes:")
    payment_modes = order_data['payment_mode'].value_counts()
    print(payment_modes)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=payment_modes.index, y=payment_modes.values, palette="Set2", hue=payment_modes.index, dodge=False)
    plt.title("Most Preferred Payment Modes")
    plt.xlabel("Payment Mode")
    plt.ylabel("Number of Orders")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    # Time-based Analysis
    print("\nTime-based analysis :- \n")
    print("Month Wise Total Sales:")
    order_data['month'] = order_data['order_date'].dt.to_period('M')
    month_wise_sales = order_data.groupby('month')['total_price'].sum()
    print(month_wise_sales)

    plt.figure(figsize=(12, 6))
    month_wise_sales.plot(kind='bar', color='teal')
    plt.title("Month Wise Total Sales")
    plt.xlabel("Month")
    plt.ylabel("Total Sales")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Month and Year Wise Total Sales:")
    order_data['year_month'] = order_data['order_date'].dt.to_period('M')
    year_month_wise_sales = order_data.groupby('year_month')['total_price'].sum()
    print(year_month_wise_sales)

    plt.figure(figsize=(12, 6))
    year_month_wise_sales.plot(kind='line', marker='o', color='purple')
    plt.title("Month and Year Wise Total Sales")
    plt.xlabel("Month and Year")
    plt.ylabel("Total Sales")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

    print("Peak Order Date:")
    peak_order_date = order_data.groupby('order_date').size().idxmax()
    print("Peak Order Date: ", peak_order_date)

    # Geographical Analysis
    print("\nGeographical analysis :- \n")
    print("Distribution of Customers Across Different Cities:")
    city_distribution = customer_data['city'].value_counts()
    print(city_distribution)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=city_distribution.index, y=city_distribution.values, palette="cubehelix", hue=city_distribution.index, dodge=False)
    plt.title("Distribution of Customers Across Different Cities")
    plt.xlabel("City")
    plt.ylabel("Number of Customers")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    print("Products or Categories More Popular in Specific Cities:")
    city_product_popularity = order_data.merge(customer_data, on='customer_id').merge(product_data, on='product_id')
    city_category_popularity = city_product_popularity.groupby(['city', 'category']).size().unstack().fillna(0)
    print(city_category_popularity)

    plt.figure(figsize=(12, 8))
    sns.heatmap(city_category_popularity, cmap="YlGnBu", annot=True, fmt='g')
    plt.title("Products or Categories Popular in Cities")
    plt.xlabel("Category")
    plt.ylabel("City")
    plt.tight_layout()
    plt.show()

    # Product Performance
    print("\nProduct performance :- \n")
    print("Top 10 Best-Selling Products:")
    best_selling_products = order_data.groupby('product_id')['quantity'].sum().nlargest(10)
    print(best_selling_products)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=best_selling_products.index, y=best_selling_products.values, palette="rocket", hue=best_selling_products.index, dodge=False)
    plt.title("Top 10 Best-Selling Products")
    plt.xlabel("Product ID")
    plt.ylabel("Quantity Sold")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    print("Top 10 Slow-Moving Products Based on Low Sales:")
    slow_moving_products = order_data.groupby('product_id')['quantity'].sum().nsmallest(10)
    print(slow_moving_products)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=slow_moving_products.index, y=slow_moving_products.values, palette="magma", hue=slow_moving_products.index, dodge=False)
    plt.title("Top 10 Slow-Moving Products")
    plt.xlabel("Product ID")
    plt.ylabel("Quantity Sold")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

    # Payment Analysis
    print("\nPayment analysis :- \n")
    print("Successful and Pending Payments Order Counts:")
    payment_status_counts = order_data['order_status'].value_counts()
    print(payment_status_counts)

    plt.figure(figsize=(10, 6))
    sns.barplot(x=payment_status_counts.index, y=payment_status_counts.values, palette="pastel", hue=payment_status_counts.index, dodge=False)
    plt.title("Successful and Pending Payments Order Counts")
    plt.xlabel("Order Status")
    plt.ylabel("Number of Orders")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.legend([],[], frameon=False)  # Disable legend
    plt.show()

# Main function to run the analysis
def main():
    customer_data, product_data, order_data = fetch_data()
    customer_data, product_data, order_data = clean_data(customer_data, product_data, order_data)
    perform_eda(customer_data, product_data, order_data)

if __name__ == "__main__":
    main()
```
